# **=============== DATA PRE-PROCESSING ===============**

```python
#@title IMPORT LIBRARY
!pip install sastrawi
!pip install tabulate
from tabulate import tabulate
!pip install vaderSentiment
from tabulate import tabulate
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

#Import library yang dibutuhkan
import csv
import nltk
import re
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemover, ArrayDictionary
import string
```

```python
#@title FILTERING TEXT
#Proses filter
def filtering_text(text):
    # mengubah tweet menjadi huruf kecil
    text = text.lower()
    # menghilangkan url
    text = re.sub(r'https?:\/\/\S+','',text)
    # menghilangkan mention, link, hastag
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    #menghilangkan karakter byte (b')
    text = re.sub(r'(b\'{1,2})',"", text)
    # menghilangkan yang bukan huruf
    text = re.sub('[^a-zA-Z]', ' ', text)
    # menghilangkan digit angka
    text = re.sub(r'\d+', '', text)
    #menghilangkan tanda baca
    text = text.translate(str.maketrans("","",string.punctuation))
    # menghilangkan whitespace berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['filtered'] = df['full_text'].apply(filtering_text)
print(df['filtered'].head())
```

```python
#@title STOPWORDS & STEMMING
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemover import StopWordRemover

def stop_stem(text, additional_stopwords=None):
    # Load stopwords dari file
    with open('/content/kamus_updated.txt') as kamus:
        word = kamus.readlines()
        list_stopword = [line.replace('\n', "") for line in word]

    # Tambahkan kata-kata tambahan ke dalam stopwords
    if additional_stopwords:
        list_stopword.extend(additional_stopwords)

    dictionary = ArrayDictionary(list_stopword)
    stopword = StopWordRemover(dictionary)
    text = stopword.remove(text)

    # Stemming
    factory_stemmer = StemmerFactory()
    stemmer = factory_stemmer.create_stemmer()
    text = stemmer.stem(text)

    return text

# Daftar kata tambahan untuk dihilangkan
additional_stopwords = ["ransomware", "keyword2", "keyword3"]  # Ganti dengan kata yang ingin dihilangkan

df['cleaned'] = df['filtered'].apply(lambda x: stop_stem(x, additional_stopwords))

print(df['cleaned'].head())
```

```python
#@title TOKENIZATION
nltk.download('punkt')

def word_tokenize_wrapper(text):
    return word_tokenize(text)

df['tweet_tokens'] = df['cleaned'].apply(word_tokenize_wrapper)

print(df['tweet_tokens'].head())
```

```python
#@title NORMALISASI KATA
kamus_normalisasi = pd.read_csv("/content/slang.csv")

kata_normalisasi_dict = {}

for index, row in kamus_normalisasi.iterrows():
    if row[0] not in kata_normalisasi_dict:
        kata_normalisasi_dict[row[0]] = row[1]

def normalisasi_kata(document):
    return [kata_normalisasi_dict[term] if term in kata_normalisasi_dict else term for term in document]

df['normalisasi'] = df['tweet_tokens'].apply(normalisasi_kata)

df['normalisasi'].head(10)
```

# **=============== DATA VISUALIZATION ===============**

```python
#@title TOP WORDS
import matplotlib.pyplot as plt
from collections import Counter

# Flatten the list of tokenized and normalized words
all_words = [word for tokens in df['normalisasi'] for word in tokens]

# Count the frequency of each word
word_counts = Counter(all_words)

# Get the top 8 most common words
top_words = word_counts.most_common(8)

# Separate the words and their counts for plotting
words, counts = zip(*top_words)

# Plot the top 8 words using a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(words, counts, color='black')
plt.xlabel('Frequency')
plt.title('Top 8 Words in Tweets')
plt.gca().invert_yaxis()  # To display the highest frequency at the top
plt.show()
```

```python
#@title TOP RETWEETS
# Mengurutkan DataFrame berdasarkan retweet_count secara menurun
top_retweets = df.sort_values(by='retweet_count', ascending=False).head(5)

# Menampilkan top 5 tweet dengan retweet terbanyak beserta detailnya
top_retweets_formatted = top_retweets[['username', 'retweet_count', 'full_text']]
print("Top 5 Tweet dengan Retweet Terbanyak:")
print(tabulate(top_retweets_formatted, headers='keys', tablefmt='grid', showindex=False))
```

```python
#@title HASTAGS ANALYISIS
# Fungsi untuk mengekstrak hashtag dari teks
def extract_hashtags(text):
    hashtags = re.findall(r'#(\w+)', text)
    return hashtags

# Menggabungkan semua teks dari kolom full_text menjadi satu teks panjang
all_texts = ' '.join(df['full_text'].astype(str))

# Mengekstrak semua hashtag dari teks panjang
all_hashtags = extract_hashtags(all_texts)

# Menghitung frekuensi penggunaan setiap hashtag
hashtags_freq = pd.Series(all_hashtags).value_counts().head(8)

# Menambahkan simbol # pada setiap hashtag
hashtags_freq.index = ['#' + hashtag for hashtag in hashtags_freq.index]

# Plot grafik dengan urutan dari besar ke kecil
plt.figure(figsize=(10, 6))
hashtags_freq.sort_values(ascending=False).plot(kind='barh', color='black')
plt.title('Top 8 Hashtags yang Digunakan', color='black')
plt.xlabel('Jumlah Penggunaan', color='black')
plt.ylabel('Hashtag', color='black')
plt.gca().invert_yaxis()
plt.gca().tick_params(axis='x', colors='black')
plt.gca().tick_params(axis='y', colors='black')
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.gca().spines['top'].set_color('black')
plt.gca().spines['right'].set_color('black')
plt.show()
```

```python
#@title TOP BIGRAM
# Menggabungkan teks tweet menjadi satu string
text_data = df['cleaned'].tolist()

# Membuat bigram menggunakan CountVectorizer
vectorizer = CountVectorizer(ngram_range=(2, 2))
X = vectorizer.fit_transform(text_data)

# Menghitung frekuensi bigram
bigram_freq = X.sum(axis=0).A1
bigram_names = vectorizer.get_feature_names_out()

# Membuat DataFrame dari bigram dan frekuensinya
bigram_df = pd.DataFrame({'bigram': bigram_names, 'frequency': bigram_freq})

# Mengurutkan berdasarkan frekuensi dan memilih 8 bigram teratas
top_bigrams = bigram_df.sort_values(by='frequency', ascending=False).head(8)

# Visualisasi grafik batang
plt.figure(figsize=(10, 6))
plt.barh(top_bigrams['bigram'], top_bigrams['frequency'], color='black')
plt.xlabel('Jumlah Kata')
plt.ylabel('Bigram')
plt.title('8 Bigram Terbesar')
plt.gca().invert_yaxis()
plt.show()
```

# **=============== SENTIMENT ANALYSIS ===============**

```python
#@title SENTIMENT ANALYSIS
# Inisialisasi VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Fungsi untuk menentukan sentimen
def get_sentiment(text):
    score = analyzer.polarity_scores(text)
    if score['compound'] >= 0.05:
        return 'positif'
    elif score['compound'] <= -0.05:
        return 'negatif'
    else:
        return 'netral'

# Terapkan fungsi pada kolom 'full_text' untuk menentukan sentimen
df['sentiment'] = df['full_text'].apply(get_sentiment)

# Menampilkan hasil dalam format tabel
formatted_output = df[['sentiment', 'full_text']].head(15)
print("Hasil Analisis Sentimen:")
print(tabulate(formatted_output, headers='keys', tablefmt='grid', showindex=False))
```

```python
#@title SENTIMENT DISTRIBUTION
# Count the occurrences of each sentiment
sentiment_counts = df['sentiment'].value_counts()

# Define labels and sizes for the pie chart
labels = sentiment_counts.index
sizes = sentiment_counts.values

# Define colors for each sentiment
colors = ['#ff9999','#66b3ff','#99ff99']

# Create the pie chart
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')

# Title of the pie chart
plt.title('Sentiment Distribution')

# Show the plot
plt.show()
```

```python
#@title SENTIMENT COUNTS
# Create the bar chart
fig2, ax2 = plt.subplots()
ax2.bar(sentiment_counts.index, sentiment_counts.values, color=colors)

# Set titles and labels
ax2.set_title('Sentiment Counts')
ax2.set_xlabel('Sentiment')
ax2.set_ylabel('Count')

# Show the plot
plt.show()
```
